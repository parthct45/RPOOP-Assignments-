```python
from random import randint

class Bank : 
    
    clients = []  

    def update(self, client):
        self.clients.append(client)

    def verification(self, name, acc_num):
        for i in range(len(self.clients)):
            if name in self.clients[i].account.values() and acc_num in self.clients[i].account.values():
                print()
                print("Authentication successful!")
                return self.clients[i]

                

class Client:
    # account is taken as an empty dictionary initially 
    account = {}

    def __init__(self, name, deposit):
        self.account['acc_num'] = randint(10000, 99999)
        self.account['name'] = name
        self.account['balance'] = deposit 

    def withdraw(self, amount):
        if self.account['balance'] >= amount:
            self.account['balance'] -= amount
            print()
            print("The sum of ",amount," has been withdrawn from your account balance.") 
            self.balance()
        else:
            print()
            print("Insufficient Balance ")
            self.balance()

    def deposit(self, amount):
        self.account['balance'] += amount
        print()
        print("The sum of ",amount," has been added to your account balance.")
        self.balance()

    def balance(self):
        print()
        print("Your current account balance is:",self.account['balance']) 

        


bank = Bank()
running = True
while running:
    print()
    print("Choose an option:\n1. Open new bank account\n2. Open existing bank account\n3. Exit")

    choice = int(input("1, 2 or 3: "))

    if choice == 1:
        print()
        print("Please fill in the information below.")
        print()
        client = Client(input("Name: "), int(input("Deposit amount: ")))
        bank.update(client)
        print()
        print("Account created successfully! Your account number is: ", client.account['acc_num'])
    elif choice == 2:
        print()
        print("Please enter your credentials below.")
        print()
        name = input("Name: ")
        acc_num = int(input("Account number: "))
        current_client = bank.verification(name, acc_num)
        if current_client:
            print()
            print("Welcome ",current_client.account['name'] ,"!")
            acc_open = True
            while acc_open:
                print()
                print("""Choose an option:
                
    1. Withdraw
    2. Deposit
    3. Balance
    4. Exit
                    """)
                acc_choice = int(input("1, 2, 3 or 4: "))
                if acc_choice == 1:
                    print()
                    current_client.withdraw(int(input("Withdraw amount: ")))
                elif acc_choice == 2:
                    print()
                    current_client.deposit(int(input("Deposit amount: ")))
                elif acc_choice == 3:
                    print()
                    current_client.balance()
                elif acc_choice == 4:
                    print()
                    print("Thank you for visiting!")
                    current_client = ''
                    acc_open = False
        else:
            print()
            print("Authentication failed!")
            print("Reason: account not found.")
            continue
    elif choice == 3:
        print()
        print("Thank You!")
        running = False 
```

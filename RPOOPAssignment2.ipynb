```python
import math

class Polygon:
    def __init__(self,numSides):
        self.numSides=numSides

    def shape(self):
        if self.numSides==3:
            return "Triangle"
        elif self.numSides==4:
            return "Square or Rectangle"
        else:
            return "polygon with sides more than 4"
   
    def area(self):
        pass
    def perimeter(self):
        pass 
    def getNumSides(self):
        return self.numSides

# hierarchical inheritance
class Rectangle(Polygon):
    def __init__(self,l,b):
        Polygon.__init__(self,4)
        self.length=l
        self.breadth=b
   
    def area(self):
        return self.length*self.breadth 
   
    def perimeter(self):
        return 2*(self.length+self.breadth)
   
    def get_dimension(self):
        return self.length,self.breadth
   
    def set_dimension(self,l,b):
        self.length=l
        self.breadth=b
   
    def shape(self):
        return "Rectangle"
    

class Triangle(Polygon):
    def __init__(self,a,b,c):
        Polygon.__init__(self,3)
        self.a=a
        self.b=b
        self.c=c
   
    def area(self):
        s=(self.a+self.b+self.c)/2
        #herons formula for calculating area 
        return (s*(s-self.a)*(s-self.b)*(s-self.c))**0.5 
   
    def perimeter(self):
        return self.a+self.b+self.c
   
    def get_dimension(self):
        return self.a,self.b,self.c
   
    def set_dimension(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c


# multilevel inheritance
class Square(Rectangle):
    def __init__(self,a):
        Rectangle.__init__(self,a,a)
        self.side=a
   
    def get_dimension(self):
        return self.side 
   
    def set_dimension(self,a):
        self.side=a
       
    def shape(self):
        return "Square"

# single level inheritance 

class Ellipse:
    def __init__(self,a,b):
        self.major=a
        self.minor=b

    def area(self):
        return math.pi*self.major*self.minor
    def shape(self):
        return "Ellipse"
   
   
class Circle(Ellipse):
    def __init__(self,r):
        self.r=r
        Ellipse.__init__(self,r,r)
    def shape(self):
        return "Circle"
    
#multiple inheritance
#physics teacher
class PTeacher:
    pname="Ramesh"
#maths teacher     
class MTeacher:
    mname="Suresh"

class Child(PTeacher,MTeacher):
    def __init__(self,name):
        self.name=name
    def chname(self):
        print(self.name)
    def teacher(self):
        print(self.pname,self.mname)
        
   

# object for square
sq=Square(5)
print(sq.shape())
print(sq.get_dimension())
print(sq.area())
print(sq.perimeter())
sq.set_dimension(10)
print(sq.get_dimension())
print(sq.getNumSides())
print()

#object for rectangle
rect=Rectangle(3,4)
print(rect.shape())
print(rect.get_dimension())
print(rect.area())
print(rect.perimeter())
rect.set_dimension(10,12)
print(rect.get_dimension())
print(rect.getNumSides())
print()

#object for triangle
tri=Triangle(3,4,5)
print(tri.shape())
print(tri.get_dimension())
print(tri.area())
print(tri.perimeter())
tri.set_dimension(8,6,10)
print(tri.get_dimension())
print(tri.getNumSides())
print()

#object for ellipse
ell=Ellipse(4,5)
print(ell.shape())
print(ell.area())
print()

#object for circle
cir=Circle(10)
print(cir.shape())
print(cir.area())
print()

#multiple inheritance
ch=Child("Parth")
ch.chname()
ch.teacher()
```
